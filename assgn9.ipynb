##### Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports,
##### technology, food, books, etc.).
##### 1. Convert text to lowercase and remove punctuaƟon.
##### 2. Tokenize the text into words and sentences.
##### 3. Remove stopwords (using NLTK's stopwords list).
##### 4. Display word frequency distribuƟon (excluding stopwords).

```python
import nltk
import re
import string
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
from nltk import FreqDist
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
import string
text = "Music is a universal language that expresses emotions, tells stories, and brings people together. It exists in countless forms—classical, pop, jazz, rock, and many more. Music can uplift, soothe, energize, or even heal. It's deeply connected to culture and identity across the world. Whether through instruments, vocals, or digital means, music remains a powerful force in human life.Its rhythms and melodies often speak louder than words."
text_lower = text.lower()
print(text_lower)
text_clean = text_lower.translate(str.maketrans('', '', string.punctuation))
print(text_clean)
```

```python
word_tokens = word_tokenize(text_clean)
sent_tokens = sent_tokenize(text)

print("Word Tokens:", word_tokens)
print("Sentence Tokens:", sent_tokens)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in word_tokens if w.lower() not in stop_words]
print("Filtered Words:", filtered_words)
```

```python
fd = FreqDist(filtered_words)
fd.plot(10, title="Top Words")
```

##### Q2: Stemming and LemmaƟzaƟon
##### 1. Take the tokenized words from QuesƟon 1 (aŌer stopword removal).
##### 2. Apply stemming using NLTK's PorterStemmer and LancasterStemmer.
##### 3. Apply lemmaƟzaƟon using NLTK's WordNetLemmaƟzer.
##### 4. Compare and display results of both techniques.

```python
ps = PorterStemmer()
lm = WordNetLemmatizer()
st = LancasterStemmer()
for word in filtered_words:
    print(f"{word} -> Stemmed1: {ps.stem(word)}, Stemmed2: {st.stem(word)},Lemmatized: {lm.lemmatize(word)}")
```

```python
vowel_words = [word for word in alphas if re.match(r'^[aeiouAEIOU]', word)]
print("Words starting with a vowel:", vowel_words)
```

##### Q4. Custom TokenizaƟon & Regex-based Text Cleaning
##### 1. Take original text from QuesƟon 1. 2. Write a custom tokenizaƟon funcƟon that:
##### a. Removes punctuaƟon and special symbols, but keeps contracƟons (e.g.,"isn't" should not be split into "is" and "n't").
##### b. Handles hyphenated words as a single token (e.g., "state-of-the-art" remainsa single token).
##### c. Tokenizes numbers separately but keeps decimal numbers intact (e.g., "3.14" should remain as is).

```python
def custom_tokenizer(text):
    pattern = r"\b\w+(?:-\w+)*\b|\d+\.\d+|\w+'\w+"
    return re.findall(pattern, text)
print(custom_tokenizer("State-of-the-art AI isn't easy to build. 3.14 is Pi."))
```

##### 3. Use Regex SubsƟtuƟons (re.sub) to:
##### a. Replace email addresses with '<EMAIL>' placeholder.
##### b. Replace URLs with '<URL>' placeholder.
